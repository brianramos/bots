```python
import sympy
import random
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

def number_to_base(n, base):
    """Converts a decimal number to a given base."""
    if n == 0:
        return [0]
    digits = []
    while n:
        digits.append(int(n % base))
        n //= base
    return digits[::-1]


def analyze_digit_frequency(numbers, base=60):
    all_digits = []
    for num in numbers:
        digits = number_to_base(num, base) # Call the newly defined function
        all_digits.extend([str(d) for d in digits])
    return Counter(all_digits)

# Data Generation (Matching ranges for primes and random numbers)
start = 1000
end = 5000
primes = list(sympy.primerange(start, end))
random_numbers = random.sample(range(start, end), len(primes)) # Exact same range!


# Analyze digit frequencies
prime_digit_counts = analyze_digit_frequency(primes)
random_digit_counts = analyze_digit_frequency(random_numbers)

prime_digit_counts_list = [prime_digit_counts.get(digit,0) for digit in all_possible_digits]
random_digit_counts_list = [random_digit_counts.get(digit,0) for digit in all_possible_digits]

heatmap_data_primes = np.array(prime_digit_counts_list).reshape(6,10)
heatmap_data_random = np.array(random_digit_counts_list).reshape(6,10)

# Filter primes containing '7'
primes_with_7 = [p for p in primes if '7' in "".join(map(str,number_to_base(p,60)))]

# Analyze '7' position frequency in primes containing '7'
position_counts = Counter()
max_length = 0  # Keep track of the maximum string length



for p in primes_with_7:

    base60_p = "".join(map(str, number_to_base(p, 60)))
    max_length = max(max_length, len(base60_p))
    for i, digit in enumerate(base60_p):
      if digit == '7':
        position_counts[i] += 1




# Create a matrix for the heatmap (Initialize as 0's)
heatmap_matrix = np.zeros((1, max_length))  # Adjust dimensions if needed



# Populate the heatmap matrix

for position, count in position_counts.items():

    heatmap_matrix[0,position] = count




# Visualization


plt.figure(figsize=(12, 3))  # Adjust figsize for better visualization
plt.imshow(heatmap_matrix, cmap='viridis', aspect='auto') #aspect auto important!
plt.title("Position Frequency of Digit '7' in Base-60 Primes Containing '7'")
plt.xlabel("Position of Digit '7'")
plt.yticks([]) #remove y axis since it is only 1 row
plt.colorbar(label='Frequency')
plt.show()

#Print out additional stats:

print(f"Number of Primes containing the digit '7' is {len(primes_with_7)}")
print(f"Proportion of primes containing the digit '7' is {len(primes_with_7)/len(primes)}")

#Print out the heatmap itself
#print(heatmap_matrix)
```
